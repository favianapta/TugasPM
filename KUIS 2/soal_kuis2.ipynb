````python
# Import untuk Eliminating Warning
import warnings
warnings.filterwarnings("ignore")
from sklearn.exceptions import FitFailedWarning 

import pandas as pd
import numpy as np 
from scipy.stats import mode 
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import LabelEncoder 
from sklearn.naive_bayes import MultinomialNB 
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import accuracy_score 
from sklearn.tree import DecisionTreeClassifier 
````

````python
# Membaca Data
df = pd.read_excel('data/dataset_quiz.xlsx')

# Menampilkan data teratas jumlah 20
df.head(20)
````

````python
# Mengecek jumlah kolom null
df.isnull().sum()
````

Mengisi Kolom yang nilai Nan

````python
df['id_ketua'] = df['id_ketua'].fillna(df['id_ketua'].mode()[0])
df['id_anggota1'] = df['id_anggota1'].fillna(df['id_anggota1'].mode()[0])
df['id_anggota2'] = df['id_anggota2'].fillna(df['id_anggota2'].mode()[0])
df['id_anggota3'] = df['id_anggota3'].fillna(df['id_anggota3'].mode()[0])
df['id_anggota4'] = df['id_anggota4'].fillna(df['id_anggota4'].mode()[0])
df['id_anggota5'] = df['id_anggota5'].fillna(df['id_anggota5'].mode()[0])
````

````python
# Mengecek jumlah kolom null
df.isnull().sum()
````

encoding label yang belum terencode

````python
e = LabelEncoder()

df['keterlibatan_vltr'] = e.fit_transform(df['keterlibatan_vltr'])
df['tahun'] = e.fit_transform(df['tahun'])
df['skema'] = e.fit_transform(df['skema'])
df['luaran_hki'] = e.fit_transform(df['luaran_hki'])
df['jenis_luaran_hki'] = e.fit_transform(df['jenis_luaran_hki'])
df['potensi_hki'] = e.fit_transform(df['potensi_hki'])
df['luaran_lain'] = e.fit_transform(df['luaran_lain'])
df['jenis_luaran_lain'] = e.fit_transform(df['jenis_luaran_lain'])
df['status_luaran_lain'] = e.fit_transform(df['status_luaran_lain'])
````

````python
# Mengecek hasil
df.head(20)
````

Mengecek data & menampilkan deskriipsi tabel

1. Pemilahan data training dan testing

````python
# Split data training
df_train, df_test = train_test_split(df, test_size=0.2, random_state=0)

# Mengecek ukuran data
print(f'Jumlah data asli: {df.shape[0]}')
print(f'Jumlah data train: {df_train.shape[0]}')
print(f'Jumlah data test: {df_test.shape[0]}')

# Mengecek rasio tiap tabel

print(f'Jumlah label data asli:\n{df.keterlibatan_vltr.value_counts()}')
print(f'Jumlah label data train:\n{df_train.keterlibatan_vltr.value_counts()}')
print(f'Jumlah label data test:\n{df_test.keterlibatan_vltr.value_counts()}')
````

2. Menentukan data validasi

````python
# Split data
df_train, df_unseen = train_test_split(df, test_size=0.2, random_state=0, stratify=df['keterlibatan_vltr'])

# Split lagi antara validasi dan testing
df_val, df_test = train_test_split(df_unseen, test_size=0.5, random_state=0, stratify=df_unseen['keterlibatan_vltr'])

# Mengecek masing-masing ukuran data

print(f'Jumlah data asli: {df.shape[0]}')
print(f'Jumlah data train: {df_train.shape[0]}')
print(f'Jumlah data val: {df_val.shape[0]}')
print(f'Jumlah data test: {df_test.shape[0]}')

# Mengecek rasio pada label

print(f'Jumlah label data asli:\n{df.keterlibatan_vltr.value_counts()}')
print(f'Jumlah label data train:\n{df_train.keterlibatan_vltr.value_counts()}')
print(f'Jumlah label data val:\n{df_val.keterlibatan_vltr.value_counts()}')
print(f'Jumlah label data test:\n{df_test.keterlibatan_vltr.value_counts()}')
````

3. Menentukan kolom menjadi "label"

````python
# Seleksi Fitur
X = df.iloc[:, 6:]
y = df['keterlibatan_vltr']

# Mengecek jumlah fitur dan instance
X.shape
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
````

4. Terapkan dua jenis metode klasifikasi dan lakukan perbandingan akurasi dengan wajib melibatkan variabel keterlibatan volunteer sampai variabel luaran-luaran yang dihasilkan! 

- DecisionTree

````python
# Mendefinisikan model Decision Tree Classifier
dt = DecisionTreeClassifier()

# Sesuaikan dt ke set training
dt.fit(X_train, y_train)

# Memprediksi label set test
y_pred_dt = dt.predict(X_test)

#  menghitung set accuracy
acc_dt = accuracy_score(y_test, y_pred_dt)
print("Test set accuracy: {:.2f}".format(acc_dt))
print(f"Test set accuracy: {acc_dt}")
````

- Naive Bayess

````python
# Inisiasi MultinomialNB
mnb = MultinomialNB()

# Fit model
mnb.fit(X_train, y_train)

# Prediksi dengan data training
y_pred_train = mnb.predict(X_train)

# Evaluasi akurasi data training
acc_train = accuracy_score(y_train, y_pred_train)

# Prediksi dengan data training
y_pred_test = mnb.predict(X_test)

# Evaluasi akurasi data training
acc_test = accuracy_score(y_test, y_pred_test)

# Print hasil evaluasi
print(f'Hasil akurasi data train: {acc_train}')
print(f'Hasil akurasi data test: {acc_test}')
````

- Random Forest

````python
# Mendefinisikan model Random Forest Classifier
rf = RandomForestClassifier()

# Sesuaikan rf ke set training
rf.fit(X_train, y_train)

# Memprediksi label set test
y_pred_rf = rf.predict(X_test)

#  menghitung set accuracy
acc_rf = accuracy_score(y_test, y_pred_rf)
print("Test set accuracy: {:.2f}".format(acc_rf))
print(f"Test set accuracy: {acc_rf}")
````

5. Lakukan tunning hyperparameter berdasarkan hasil klasifikasi pada nomor 3 menggunakan metode Grid Search dan Random Search! Simpulkan parameter apa yang paling baik digunakan berdasarkan metode klasifikasi yang dipilih, dan berapa akurasinya.

````python
from sklearn.model_selection import RepeatedKFold

# inisiasi repated k-fold
cv = RepeatedKFold(n_splits=4, n_repeats=3, random_state=52)
````

Grid Search

- Decission Tree

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# inisiasi model
dt = DecisionTreeClassifier()

# Definisikan hyperparameter yang akan digunakan
# sklearn menerima dalam bentuk dictionary
# nama hyperparamater HARUS SESUAI dengan dokumentasi sklearn
params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': list(range(5,11)),
    
}

# inisiasi grid berdasarkan nilai repeated k-fold dan hyperparameter
grid = GridSearchCV(dt, param_grid=params, cv=cv)

# Fit / latih berdasarkan grid
# %timeit merupakan magic command didalam ipython notebook
# yang dapat kita gunakan untuk menghitung waktu komputasi
# cara ini cukup efektif untuk melakukan evaluasi suatu algoritma atau prosedur
%timeit grid.fit(X, y)

# Evaluasi dengan score
score = grid.score(X,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {grid.best_params_}')
print(f'Best Estimator: {grid.best_estimator_}')
print(f'Best Score: {grid.best_score_}')
````

- Random Forest

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# inisiasi model
rf = RandomForestClassifier()

# Definisikan hyperparameter yang akan digunakan
# sklearn menerima dalam bentuk dictionary
# nama hyperparamater HARUS SESUAI dengan dokumentasi sklearn
params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': list(range(5,11)),
}

# inisiasi grid berdasarkan nilai repeated k-fold dan hyperparameter
grid = GridSearchCV(rf, param_grid=params, cv=cv)

# Fit / latih berdasarkan grid
# %timeit merupakan magic command didalam ipython notebook
# yang dapat kita gunakan untuk menghitung waktu komputasi
# cara ini cukup efektif untuk melakukan evaluasi suatu algoritma atau prosedur
%timeit grid.fit(X, y)

# Evaluasi dengan score
score = grid.score(X,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {grid.best_params_}')
print(f'Best Estimator: {grid.best_estimator_}')
print(f'Best Score: {grid.best_score_}')
````
